```python
w0=0
w1=1.2
w2=0.6
z=w0*1+w1*X_df['x1']+w2*X_df['x2']
print(z)
yp=[]
for i in range(4):
  if z[i]>0:
    yp.append(1)
  else:
    yp.append(0)
print(yp)
```

```python
error = np.array(y) - np.array(yp)
print(error)
```

```python
e_update1=w0+0.5*(error[3])*1
w0=e_update1
e_update2=w1+0.5*(error[3])*0.1
w1=e_update2
e_update3=w2+0.5*(error[3])*0.1
w2=e_update3
print(w0)
print(w1)
print(w2)
```

```python
x0 = [1, 1, 1, 1]
x1 = [0.4, 0.3, 0.2, 0.1]
x2 = [0.5, 0.8, 0.9, 0.1]
y = [1, 1, 1, 0]

w0 = 0
w1 = 1.2
w2 = 0.6

learning_rate = 0.5

epochs = 1

print("Initial Weights: w0={}, w1={}, w2={}".format(w0, w1, w2))

for epoch in range(epochs):
    print("\nEpoch:", epoch + 1)
    for i in range(len(x1)):
        net_input = (x0[i] * w0) + (x1[i] * w1) + (x2[i] * w2)

        if net_input >= 0:
            prediction = 1
        else:
            prediction = 0

        error = y[i] - prediction

        w0 = w0 + learning_rate * error * x0[i]
        w1 = w1 + learning_rate * error * x1[i]
        w2 = w2 + learning_rate * error * x2[i]


        print("  Input: x1={}, x2={}, Target: {}, Prediction: {}, Error: {}".format(x1[i], x2[i], y[i], prediction, error))
        print("  Updated Weights: w0={:.4f}, w1={:.4f}, w2={:.4f}".format(w0, w1, w2))

print("\nTesting the trained perceptron:")
for i in range(len(x1)):
    net_input = (x0[i] * w0) + (x1[i] * w1) + (x2[i] * w2)
    if net_input >= 0:
        prediction = 1
    else:
        prediction = 0
    print("  Input: x1={}, x2={}, Predicted Output: {}".format(x1[i], x2[i], prediction))
    print("\nFinal Weights: w0={:.4f}, w1={:.4f}, w2={:.4f}".format(w0, w1, w2))
```
